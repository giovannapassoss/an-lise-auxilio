```python
import pandas as pd 
import  pdfplumber
import matplotlib.pyplot as plt 
```

```python
#Transformando o arquivo csv em data frame 

df = pd.read_csv("C:\\Users\\rober\\Downloads\\resultados_Transporte.csv", encoding='latin-1', sep= ";")
```

```python
#lendo o Data Frame 

df.head(66)
```

```python
#criando um data frame  ordenando a faixa auxilio do maior para o menor 

df_sorted = df.sort_values(by="Faixa do Auxílio", ascending=False)
```

```python
# lendo o dataframe 
df_sorted.head(66)
```

```python
# Aqui fiz um data frame que está filtrando  apenas pessoas que escolheram a faixa 3 do auxilio 

df_faixa3 = df[df["Faixa do Auxílio"] == 3].sort_values(by="Faixa do Auxílio", ascending=True)
```

```python
df_faixa3.head(50)
```

```python
# Dataframe  escolhendo apenas as pessoas que escolheram a faixa 3 e ordenando a pontuação do maior para o menor 

df_faixa3_sorted = df[df["Faixa do Auxílio"] == 3].sort_values(by="Pontuação Final", ascending=False)
```

```python
#lendo o dataframe 

df_faixa3_sorted.head(66)
```

```python

#Usa str.contains para garantir que o nome seja encontrado de forma flexível

df_giovanna = df_faixa3_sorted[df_faixa3_sorted["Nome"].str.contains("Giovanna Oliveira Pinto Passos da Silva", case=False)]
```

```python

if not df_giovanna.empty:
    pontuacao_giovanna = df_giovanna["Pontuação Final"].values[0]
    pessoas_na_frente = len(df_faixa3_sorted[df_faixa3_sorted["Pontuação Final"] > pontuacao_giovanna])
    print(f"Existe : {pessoas_na_frente} pessoas na sua frente")
else:
    print("Giovanna não foi encontrada na faixa 3.")
```

```python
#gerando gráfico mostrando onde está a minha posição na lista 
pessoas_na_frente = 27
posicao_giovanna = pessoas_na_frente + 1
total_vagas = 50


plt.figure(figsize=(6, 4))
plt.bar(['Pessoas na frente', 'Posição de Giovanna'], [pessoas_na_frente, posicao_giovanna], color=['blue', 'green'])

plt.title('Posição de Giovanna para  receber o auxílio')
plt.ylabel('Número de pessoas')
plt.ylim(0, total_vagas)


plt.show()
```
